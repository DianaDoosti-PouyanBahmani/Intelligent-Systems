# Preprocessing

```python
# --- Check for duplicate rows ---
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

if num_duplicates > 0:
    print("\nExample of duplicate rows:")
    print(df[df.duplicated()].head())

# --- Overall summary ---
total_missing = df.isnull().sum().sum()
print(f"\nTotal missing values in the dataset: {total_missing}")

if total_missing > 0:
  print("\nMissing values per column:")
  print(df.isnull().sum())
  print("\nPercentage of missing data per column:")
  print((df.isnull().mean() * 100).round(2))

```

```python
# --- Apply One-Hot Encoding ---
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# --- Check results ---
print("\nEncoded dataset shape:", df_encoded.shape)
print("\nExample of encoded columns:\n", df_encoded.head())
```

```python
# store outlier flags for each col
outlier_flags = pd.DataFrame(False, index=df_encoded.index, columns=numeric_cols)

IQR_MULT = 1.5  # or 2.0 for more forgiving

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - IQR_MULT * IQR
    upper = Q3 + IQR_MULT * IQR
    outlier_flags[col] = (df[col] < lower) | (df[col] > upper)

# how many columns are outliers per row?
outlier_count = outlier_flags.sum(axis=1)

# drop only rows that are outliers in >= 2 columns
df_new = df_encoded[outlier_count < 2].copy()

print("Original shape:", df_encoded.shape)
print("Soft-cleaned shape:", df_new.shape)
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

TARGET_COL = "price"

# ---- 1. Separate features and target ----
X = df_new.drop(columns=[TARGET_COL])
y = df_new[TARGET_COL]

# ---- 2. Split into Train and Test sets ----
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% for testing
    random_state=42,    # ensures reproducibility
    shuffle=True
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

# ---- 3. Normalize numeric features (Standardization) ----
scaler = StandardScaler()

# Fit only on training data to avoid data leakage
scaler.fit(X_train)

# Transform both training and test data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Normalization complete.")
```

MLP (Multi-Layer Perceptron)

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt

# Prepare data (using PCA-transformed data)
X_train_tensor = torch.tensor(X_train_pca, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_pca, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)

# Define the MLP model
class MLP(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)  # First hidden layer
        self.fc2 = nn.Linear(128, 64)         # Second hidden layer
        self.fc3 = nn.Linear(64, output_dim)  # Output layer
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))  # Apply ReLU after first layer
        x = self.relu(self.fc2(x))  # Apply ReLU after second layer
        x = self.fc3(x)             # Output layer without activation
        return x

# Hyperparameters
input_dim = X_train_pca.shape[1]
output_dim = 1
batch_size = 10  # Batch size for mini-batch training
num_epochs = 500
learning_rate = 0.01

# Create DataLoader for mini-batch processing
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Model, Loss function, and Optimizer
model = MLP(input_dim, output_dim)  # No hidden_dim needed here anymore
criterion = nn.MSELoss()  # Mean Squared Error for regression
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Early Stopping Parameters
patience = 10  # Number of epochs with no improvement before stopping
best_loss = np.inf
patience_counter = 0

# Initialize loss lists for plotting
train_losses = []  # To store training loss per epoch
val_losses = []    # To store validation loss per epoch

# Training the model with Early Stopping and Mini-batch training
for epoch in range(num_epochs):
    model.train()  # Set model to training mode
    running_loss = 0.0

    # Iterate over mini-batches
    for batch_X, batch_y in train_loader:
        # Forward pass
        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Average loss for the epoch
    avg_train_loss = running_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # Check validation loss (on the test set)
    model.eval()  # Set model to evaluation mode
    with torch.no_grad():  # Turn off gradient calculation
        y_pred = model(X_test_tensor)
        val_loss = criterion(y_pred, y_test_tensor)
        val_losses.append(val_loss.item())

    # Print training and validation loss every 10 epochs
    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Training Loss: {avg_train_loss:.4f}, Validation Loss: {val_loss.item():.4f}")

    # Early stopping logic
    if val_loss < best_loss:
        best_loss = val_loss
        patience_counter = 0  # Reset the patience counter
    else:
        patience_counter += 1

    if patience_counter >= patience:
        print(f"Early stopping at epoch {epoch+1} due to no improvement.")
        break

# Evaluate the final model
model.eval()  # Set model to evaluation mode
with torch.no_grad():
    y_pred = model(X_test_tensor)

# Evaluate model performance
mse = mean_squared_error(y_test_tensor, y_pred)
r2 = r2_score(y_test_tensor, y_pred)

print(f"Final Test Set - MSE: {mse:.3f}, R²: {r2:.3f}")

# --- Plot Training and Validation Loss over Epochs ---
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Training Loss', color='blue')
plt.plot(val_losses, label='Validation Loss', color='red')
plt.title('Training and Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

# MLP dimension reduction

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt

X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
X_test_tensor  = torch.tensor(X_test_scaled,  dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values if hasattr(y_train, "values") else y_train,
                              dtype=torch.float32).view(-1, 1)
y_test_tensor  = torch.tensor(y_test.values if hasattr(y_test, "values") else y_test,
                              dtype=torch.float32).view(-1, 1)

input_dim = X_train_tensor.shape[1]
latent_dim = 4  # <-- bottleneck = 4 learned features

# Autoencoder: learns a 4-D nonlinear representation

class Autoencoder(nn.Module):
    def __init__(self, input_dim, latent_dim=4):
        super().__init__()
        # Encoder
        self.enc_fc1 = nn.Linear(input_dim, 128)
        self.enc_fc2 = nn.Linear(128, 32)
        self.bottleneck = nn.Linear(32, latent_dim)   # <-- 4 neurons
        # Decoder
        self.dec_fc1 = nn.Linear(latent_dim, 32)
        self.dec_fc2 = nn.Linear(32, 128)
        self.out = nn.Linear(128, input_dim)
        self.relu = nn.ReLU()

    def encode(self, x):
        x = self.relu(self.enc_fc1(x))
        x = self.relu(self.enc_fc2(x))
        z = self.bottleneck(x)
        return z

    def decode(self, z):
        x = self.relu(self.dec_fc1(z))
        x = self.relu(self.dec_fc2(x))
        x = self.out(x)  # reconstruction (no activation)
        return x

    def forward(self, x):
        z = self.encode(x)
        x_hat = self.decode(z)
        return x_hat

# Hyperparameters (kept similar)
batch_size = 30
num_epochs_ae = 400
lr_ae = 1e-3

ae = Autoencoder(input_dim=input_dim, latent_dim=latent_dim)
ae_criterion = nn.MSELoss()              # reconstruction loss
ae_optimizer = optim.Adam(ae.parameters(), lr=lr_ae)

ae_train_ds = TensorDataset(X_train_tensor, X_train_tensor)  # input -> target = input
ae_train_loader = DataLoader(ae_train_ds, batch_size=batch_size, shuffle=True)

# Early stopping for AE
patience = 10
best_val = np.inf
patience_counter = 0

ae_train_losses, ae_val_losses = [], []

for epoch in range(num_epochs_ae):
    ae.train()
    running = 0.0
    for xb, _ in ae_train_loader:
        xhat = ae(xb)
        loss = ae_criterion(xhat, xb)
        ae_optimizer.zero_grad()
        loss.backward()
        ae_optimizer.step()
        running += loss.item()
    train_loss = running / len(ae_train_loader)
    ae_train_losses.append(train_loss)

    # validation on X_test
    ae.eval()
    with torch.no_grad():
        xhat_val = ae(X_test_tensor)
        val_loss = ae_criterion(xhat_val, X_test_tensor).item()
        ae_val_losses.append(val_loss)

    if (epoch + 1) % 10 == 0:
        print(f"[AE] Epoch {epoch+1}/{num_epochs_ae} | train={train_loss:.4f} | val={val_loss:.4f}")

    # early stopping
    if val_loss < best_val:
        best_val = val_loss
        best_state = {k: v.cpu().clone() for k, v in ae.state_dict().items()}
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"[AE] Early stopping at epoch {epoch+1}.")
            break

# Load best AE weights
ae.load_state_dict(best_state)

# Plot AE losses
plt.figure(figsize=(8,5))
plt.plot(ae_train_losses, label="AE Train Loss")
plt.plot(ae_val_losses, label="AE Val Loss")
plt.title("Autoencoder Reconstruction Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.grid(True); plt.legend(); plt.show()

# Extract 4 learned features from bottleneck

ae.eval()
with torch.no_grad():
    Z_train = ae.encode(X_train_tensor).cpu().numpy()  # shape (n_train, 4)
    Z_test  = ae.encode(X_test_tensor).cpu().numpy()   # shape (n_test, 4)

# Convert to tensors
Z_train_t = torch.tensor(Z_train, dtype=torch.float32)
Z_test_t  = torch.tensor(Z_test,  dtype=torch.float32)
```

# VIF and RFE methods

VIF method


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Compute VIFs once -----
def compute_vif(df: pd.DataFrame) -> pd.DataFrame:
    # statsmodels expects a NumPy array; no constant needed for VIF
    values = df.values
    vif = [variance_inflation_factor(values, i) for i in range(values.shape[1])]
    return pd.DataFrame({"feature": df.columns, "VIF": vif}).sort_values("VIF", ascending=False)

X_vif = pd.DataFrame(X_train_scaled, columns=X_train.columns)

print("Initial VIFs:")
print(compute_vif(X_vif))

#  Iteratively drop features with VIF > threshold -----
def drop_high_vif(df, threshold=5.0, max_iter=10):
    remaining = df.copy()
    for _ in range(max_iter):
        vif_df = compute_vif(remaining)
        worst_feat, worst_vif = vif_df.iloc[0]["feature"], vif_df.iloc[0]["VIF"]
        if worst_vif <= threshold:
            return remaining, vif_df
        # drop the single worst offender and continue
        remaining = remaining.drop(columns=[worst_feat])
    return remaining, compute_vif(remaining)

X_reduced, final_vifs = drop_high_vif(X_vif, threshold=5.0)
print("\nReduced feature set after VIF filtering:")
print(X_reduced.columns.tolist())
print("\nFinal VIFs:")
print(final_vifs)
```

# Elastic Regression Model

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def run_elasticnet(Xtr, Xte, ytr, yte, name, add_scaler=False):
    steps = []
    if add_scaler:
        steps.append(("scaler", StandardScaler()))
    steps.append(("enet", ElasticNet(max_iter=10000, random_state=42)))
    pipe = Pipeline(steps)

    param_grid = {
        "enet__alpha": np.logspace(-4, 3, 20),            # strength of regularization
        "enet__l1_ratio": [0.1, 0.3, 0.5, 0.7, 0.9, 0.95, 0.99, 1.0],  # 0=L2, 1=L1
    }
    gs = GridSearchCV(
        pipe,
        param_grid=param_grid,
        scoring="neg_mean_squared_error",
        cv=5,
        n_jobs=-1,
        verbose=0,
        refit=True,
    )
    gs.fit(Xtr, np.ravel(ytr))

    y_pred = gs.predict(Xte)
    metrics = {
        "Dataset": name,
        "Best alpha": gs.best_params_["enet__alpha"],
        "Best l1_ratio": gs.best_params_["enet__l1_ratio"],
        "MSE": mean_squared_error(yte, y_pred),
        "MAE": mean_absolute_error(yte, y_pred),
        "R2": r2_score(yte, y_pred),
    }
    return metrics, gs

# ----- Run on BOTH datasets -----
# Full feature set is already scaled -> no extra scaler
scaled_metrics, scaled_gs = run_elasticnet(
    X_train_scaled, X_test_scaled, y_train, y_test, name="Scaled (no PCA)", add_scaler=False
)

# PCA features: they may not be unit-variance; add a scaler for fairness/comparability
pca_metrics, pca_gs = run_elasticnet(
    X_train_pca, X_test_pca, y_train, y_test, name="PCA features", add_scaler=True
)

# features reduced by the mlp encode/decoder
mlp_metrics, mlp_gs = run_elasticnet(
    Z_train, Z_test, y_train, y_test, name="MLP features", add_scaler=False
)

# ----- Compare results -----
comparison = pd.DataFrame([scaled_metrics, pca_metrics, mlp_metrics])[[
    "Dataset", "Best alpha", "Best l1_ratio", "MSE", "MAE", "R2"
]]
print(comparison.to_string(index=False))

# Optional: peek at coefficient counts
coef_scaled = scaled_gs.best_estimator_.named_steps["enet"].coef_
coef_pca    = pca_gs.best_estimator_.named_steps["enet"].coef_
coef_mlp    = mlp_gs.best_estimator_.named_steps["enet"].coef_
print(f"\nNon-zero coefficients -> Scaled: {(coef_scaled!=0).sum()} / {len(coef_scaled)}, "
      f"PCA: {(coef_pca!=0).sum()} / {len(coef_pca)}",
      f"MLP: {(coef_pca!=0).sum()} / {len(coef_mlp)}")
```
